## 05-03. 클릭 트리거 생성시 유의 사항
- **웹 사이트에서 같은 HTML요소라고 보여도, 요소 안의 작은 세부적인 요소 (혹은 클래스)가 존재할 수 있음**

<br>

### 찜하기

<img src='image/GTM/050112.png' width=300>

<br>

- 위 찜하기 버튼은 하나의 Click Classes를 가진 것 처럼 보이나,

    버튼(=박스)안의 하트 이모티콘을 클릭하여 찜하기가 되었을 때, 클릭한 요소의 `Click Classes`는 `_wish_button_icon wish-icon-s2025102804901b6aad302 im-icon im-ico-like`

    버튼을 클릭하여 찜하기가 되었을 떄는, `_wish_button btn like_box`

    이모티콘을 클릭하여 찜하기가 해제 되었을 때는, `_wish_button_icon wish-icon-s2025102804901b6aad302 im-icon im-ico-li" +"ked`

    버튼을 클릭하여 찜하기가 해제 되었을 때는, `_wish_button btn like_box active`

    $\rightarrow$ **측정하고 싶은 이벤트에 대한 모든 경우의 수를 트리거로 지정해 주어야 함**

    $\rightarrow$ **많은 디버깅과 테스트 필요**

### 트리거 생성 조건

| 트리거 유형                                   | 내부 이벤트      | 실행 타이밍                                        | DOM 접근 가능  | 추천 사용 상황                         |
| ---------------------------------------- | ----------- | --------------------------------------------- | ---------- | -------------------------------- |
| **Page View – Container Loaded**         | `gtm.js`    | GTM 컨테이너 코드가 로드된 직후 (가장 빠름)                   | ❌          | GA4 기본 설정, 단순 픽셀 태그              |
| **Page View – DOM Ready**                | `gtm.dom`   | HTML 구조 파싱 완료 후 (`DOMContentLoaded`)          | ✅          | 버튼, 텍스트, data-속성 등 DOM 요소를 읽는 태그 |
| **Page View – Window Loaded**            | `gtm.load`  | 모든 이미지/JS/CSS 리소스까지 다 로드된 후 (`window.onload`) | ✅          | 전환 스크립트, 이미지 픽셀, 리타게팅 태그 등       |
| **Page View – Page View (Custom Event)** | `page_view` | `dataLayer.push({event:'page_view'})` 발생 시    | ⚙️ depends | SPA 등 수동 페이지 전환 감지               |

<br>

| 상황                                | 선택                                                |
| --------------------------------- | ------------------------------------------------- |
| **GA4 Configuration / 기본 분석 태그**  | ✅ *Page View → Container Loaded*                  |
| **상품 ID, 가격 등 DOM 요소 읽어야 할 때**    | ✅ *Page View → DOM Ready*                         |
| **이미지 기반 리타게팅 픽셀, conversion 태그** | ✅ *Page View → Window Loaded*                     |
| **SPA (Imweb, React 등)**          | ✅ *Custom Event → page_view (History Change로 푸시)* |
