### 항목 ID
- 상품의 고유한 ID값을 의미
  - 이름이 같더라도 ID가 다른 상품이 있을 수 있기 때문에 반드시 필요
  
  <img src='https://lh7-us.googleusercontent.com/xZ2a2D_GQivXsoafSyWznOb_45HVFkqtINMMjkNP0b3MkZYACf-QT9b84_MhEgfNXElBe8zbH6qtLAxRA_ggYMqEBaw2HPyIfY-t6F8NmLPd6GCaSJ1FZmqRr6KUxgHqOdH0EWc-pYSowzULuQGUJkM' width='500'>

  - 어떤 쇼핑몰이든 모든 상품마다 ID값이 존재. 
    - 예시 스크립트에는 SKU_로 상품의 식별 ID가 있으나, 카페24는 콘솔창에서 제품의 ID를 검색하면 223과 같이 숫자가 나오는것을 볼 수 있음
    - 카페24는 별도의 형식없이, 제품을 업로드하는대로 제품의 고유 식별자를 숫자대로 붙이며, 제품 ID가 어떤 형식 없이 위와같이 숫자로만 이루어져있다면 당연히 GA4로도 아래와 같이 데이터가 나타남

  <img src='https://lh7-us.googleusercontent.com/HmTlahwaaPg0RoNdtWZdjjZyg-B8jZgzlOsxBwP-KOUQvnQsWbMzhJ0IbNLfg817FdryRJvIM11TG1krjjx_H9Ip_ZBGPkq94EwmkmUL5EfgBD6Gf7G7agqy1il9eWxO3i70f9qcMWWythiv98iFdVU' width='500'>

### 항목 카테고리
- 항목 카테고리는 말그대로 특정 상품의 카테고리를 의미하며, `item` 객체 안에 있는 `Key` 중에서, `item_category`이라는 Key의 Value값이 항목 카테고리를 차지
  
  <img src='https://lh7-us.googleusercontent.com/NQCELSXivoeTUXbpiHvDSuRviO_EWzaENDD-tfP7m0g6zVmlO2j_9dOQSYeknI5yX-Bn5atG1a90S2f7_HT-9DKVE8IJ84Jsf_O-HQ_ix8g0yyMAfZATCbiMIQ4JYv3x-ZVGmidR-qMQEEzIdypQBuY' width='500'>

<br>

### 항목 세부 카테고리
- 상품 카테고리의 숫자가 많아질수록 조금 더 세부적인 카테고리 값을 넣도록 일반적으로 설계
  
  <img src='https://lh7-us.googleusercontent.com/MK5nNVvFc8o52mYSNNo7Ly2dx3K7yVzfFhEpWsHW68HGojJi2RR8PKdB41vvQtIMFCYACpnTR7clutds4OH1Ld3I8h1sE8Bulk2qdznoJACfRhK0GOKUuQoAqCbNJtS3Dz599NVJVJYvJnc5UrC4QsU' width='500'>

- 세부적인 카테고리가 없다면 굳이 `item_category5`까지 값을 모두 넣어줄 필요가 없으며, 필요에 따라 자유롭게 카테고리에 있는 슬롯들을 사용
  - 최대 5개까지만 가능
- `item_category` 매개변수를 받아오는 측정기준의 이름은 항목 카테고리 
  - `item_category 2`의 매개변수를 받아오는 측정 기준의 이름은 상품 카테고리 2 
  - 또 상품 카테고리 3,4,5 처럼 제일 최초의 `item_category`를 제외하고는 모두 다 상품 카테고리를 사용하여 별도로 관리할 때에도 있음

<br>

### 항목 대안
- **항목대안은 제품의 옵션명을 의미**
- item 객체 안에 있는 Key 중에서, `variant` 라는 Key의 Value값이 항목 이름을 채움
- 제품의 옵션명이 따로 없는 단일 제품이라면 굳이 전송하지 않음

### 상품 브랜드
- 상품의 브랜드명
- 1개의 브랜드에서 다양한 제품을 판매하는 경우에는 브랜드 이름이 유일하게 한개로 들어가지만, 
  
  다양한 제품을 판매하는 유통 쇼핑몰의 경우 브랜드가 여러개일 수 있음 (ex. 무신사, 쿠팡 등) 
- item 객체 안에 있는 Key 중에서,  `brand`라는 Key의 Value값이 항목 이름을 채움

## `표준 리포트` > `GA4 전자상거래 구매` > `측정 항목`

<img src='https://lh7-us.googleusercontent.com/BR6y4JA7EdFGsgckgZNObQjfjXasL2KNfBOePTcbOiQbiTELpPW_VxLls-JBJJPnUvQkxCXsani9AoiYIPFEFKCBW3V6KD4FHCK9nTYzyUvSXseWjqh5_Hd0HeX12NC9NjQXCFj8CHnnq_F4-uABLz8' width='600'>

- 전자상거래 구매 리포트의 4가지의 측정 항목

<br>

1) 조회된 상품 : 특정 상품의 상세페이지 조회 수를 의미
- 해당 측정 항목을 보기 위해서라면 상세페이지 조회가 발생할 때, `view_item` 이벤트를 전송

2) 장바구니에 추가된 상품 : 특정 상품의 장바구니 담기 수
- 해당 측정 항목을 보기 위해서라면 장바구니 담기가 발생할 때, `add_to_cart` 이벤트를 전송

3) 구매한 상품 : 특정 상품의 구매 횟수를 의미
- 해당 측정 항목을 보기 위해서라면 구매가 발생할 때, `purchase` 이벤트를 전송

4) 상품 수익 : 특정 상품의 수익을 의미
- Item 객체에서 `price`와 `quantity`를 곱한값으로 계산이 처리

  <img src='https://lh7-us.googleusercontent.com/tO23O5ksA6APV1z2EeBbChp2J9WeIgYjJpS7BZAEGirFfR3bYoOxlg3Bb5NLR0_KU5Bm2sR0NPc63Xdz6aQ1WYN4AVlBEYN-Ecj7wIom3xjK1CqoOyvoRlA3v5xrW5CN-qvPCx2PXkMhs1LxHBjMqB4' width='500'> 

- `discount` 할인 금액은 상품 수익을 계산할때에는 별도로 빼지않으며, 최종 주문 자체에 대한 수익을 계산할 때 사용

  **이처럼 items 배열에 어떤 Key값과 Value값을 전송하느냐에 따라서, 또는 어떤 이벤트를 어떤 시점에 보내느냐에 따라서, GA4 전자상거래 리포트에서는 쇼핑몰에 필요한 다양한 데이터를 볼 수 있음**

<br>
