## GA4 설치 점검 방법

<br>

### 1. GA 실시간 보고서를 통한 확인
- GA태그가 정상적으로 설치되었다면, 사이트에 접속했을 때, 해당 보고서에 트래픽이 기록
    - 웹 사이트 데이터를 확인하기 위해서는 브라우저에서 제공하는 시크릿 모드를 활용하는 것이 권장
    - 시크릿 모드는 웹 브라우저에서 해당 기기에 쿠키와 웹 서핑 기록이 남지 않음

<br>

### 2. 외부 툴 이용
- 사이트의 HTML태그를 기반으로 GA4 설치 여부 확인
- 크롬 확장 프로그램 **`Tag Assistant Companion`**, **`Omnibug`**
