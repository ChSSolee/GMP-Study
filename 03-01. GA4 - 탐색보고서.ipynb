2. 설정
- 변수 영역에서 가져온 항목들을 기반으로 보고서를 생성

        - 데이터를 익숙한 테이블 형태로 출력되는 보고서
        - 막대 그래프, 원형 차트, 선 차트, 분산형 카드, 지역 지도 등 다양한 시각화 스타일을 적용 가능

        - 공통된 특성을 가진 사용자 그룹의 행동과 실적에서 유용한 정보를 얻을 수 있는 보고서
    3. **유입경로 탐색 분석**
        - 퍼널 형태로 시각화하여 사용자가 사이트 또는 앱에서 작업을 처리하기 위해 조치를 시각적으로 볼 수 있음 
            
            $\rightarrow$ 사용자 경험을 최적화 하고 실적이 우수한, 또는 저조한 잠재조객을 파악하는 방법을 확인 
    4. **세그먼트 중복 분석**
        - 다양한 사용자 세그먼트 간의 관계를 확인할 수 있는 보고서
    5. **경로 탐색 분석**
        - 사용자가 웹사이트 및 앱과 상호작용을 할 떄 사용하는 경로를 시각적으로 확인할 수 있는 보고서
    6. **사용자 전체 기간**
        - 전체 기간에 걸쳐 고객의 가치 및 사용자 행동을 분석하는 보고서
    7. **사용자 탐색 분석**
        - 세그먼트를 구성하는 사용자를 검토할 수 있는 보고서

## 03-01. 탐색 분석 보고서 세팅
- 데모 계정에서 세팅해볼 수 있는 기법은 자유형식, 동질 집단, 유입 경로, 세그먼트, 경로 탐색, 사요자 전체 기간 총 6가지

<br>

### 자유 형식
- 자유형식을 이용하는 방법은 크게 두 가지로 분류
1. **변수 탬에서 사용할 것 같은 세그먼트, 측정기준, 측정항목을 가져오는 것**
2. **탭 설정 단계에서 변수 탭에서 가져온 것들을 의동한 방식에 맞게, '세그먼트 비교, 행, 열, 값'에 배치하는 것**
    - '행, 열'에는 측정기준을, '값'에는 측정항목을 배치

<img src='image\GTM\030103.png' width=700>

- '시작 행' : 출력된 전체 데이터에서 몇 번째 열을 시작 행으로 설정할지 정하는 옵션
    - 너무 많은 수치를 보이는 항목을 제외하거나, 의미있는 비교를 할 수 있는 경우에 사용

        $\rightarrow$ 몇 행 부터 보여줄지 선택

- '행 표시' : 한 화면에 몇 개의 행 데이터를 출력할 것인가를 의미 (최대 500)
- '중첩된 행' : 측정기준이 2개 이상일 때, 계층적 확인 가능
- '열' : 측정기준 중 하나를 열에 배치하면 새로운 축에도 측정기준이 추가되어 데이터가 더 세부적으로 나뉨
- '값' : 셀 유형 선택
    - 셀 유형으로는 막대 차트, 일반 텍스트, 히트맵 3가지

<br>

#### 필터
- 변수 탭에서 가져온 측정기준과 측정항목 값을 기준으로 원하는 데이터만 남길 수 있게 하는 기능
- 필요에 따라 여러 개의 필터를 설정할 수 있음

<br>

### 동질 집단 탐색 분석
- GA4의 동질 집단 탐색 분석은 코호트 분석과 동일
- 동질 집단 탐색 분석은 사용자들을 특정 기간에 걸쳐 동일한 경험을 한 그룹(코호트)로 나누고, 이들 그룹의 행동 변화를 시간에 따라 추적하는 분석 방법

    $\rightarrow$ 제품이나 서비스의 장기적인 가치를 평가하고, 사용자 이탈 패턴을 파악하는 등의 인사이트 얻을 수 있음

    - **이커머스 웹사이트에서 코호트 분석을 실식하여, 지역별로 가입한 신규 사용자의 행동 패턴을 이해하는 데 주로 활용되며, 서비스를 제공하는 기업의 경우, 사용자들의 잔존율 (Retention) 데이터를 확인하는 데 활용**

<img src ='image\GTM\030104.png' width=560>

<br>

- 동질 집단 탐색 분석의 보고서에서 세로축은 분석하고자 하는 사용자 그룹
    
    보통 같은 시간대에 웹사이트에 가입하거나 어떤 중요한 행동(첫 구매 등)을 한 사용자들로 구성
    
    **세로축을 따라 내려가면, 다른 시간대에 행동을 시작한 사용자 그룹**

- 가로축은 시간의 흐름을 의미, **왼쪽에서 오른쪽으로 이동하면서 사용자가 그룹에 속한 후 지난 시간을 볼 수 있음**

<br>

- **'동질 집단 포함'** : **동질 집단 포함 기준을 무엇으로 규정할지를 정하는 것**
    - 첫 번째 터치 : 웹이나 앱에 처음으로 들어온 순간
    - 기타 :생성한 세부 이벤트
        - 예) `view_item` 선택 $\rightarrow$ 해당 이벤트를 발동시킨 사용자를 동질 집단으로 인식

- **재방문 기준** : **사용자가 언제 다시 웹사이트나 앱을 방문하였다고 규정할지를 정하는 기준**
    - 예) 우리 사이트에 들어온 사용자가 오늘 이벤트를 생성하였다면, 그 사용자는 재방문?
- **동질 집단 세분화** : **코호트 분석에서 기간을 어떤 단위로 볼지를 설정**

<img src='image\GTM\030105.png' width=200>

<br>

- 동질 집단 탐색 분석에는 일반 (Standard), 최근 (Rolling), 누적 (Cumulative) 3가지 존재
    
    - **일반** : 세로축의 기간에 처음 접속한 사용자에 대해 세팅한 조건의 결과로 데이터르 보여줌
        - 첫 주차에 처음 유입한 사용자가 +1 주차에 재방문하지 않았더라도, +2주차에 재방문했다면, 이 사용자에 대한 데이터는 카운팅
    
    - **최근** : 최초 시점부터 해당 기간까지 세분화 기준 기간 (일, 주, 월)마다 재방문 기준을 모두 달성한 사용자의 데이터만을 보여주는 보수적인 계산법
        - **첫 주차에 처음 접속한 사용자가 +1주 차에 재방문하지 않고, +2주 차에 재방문했다면, 이 사용자에 대한 데이터는 +1주차 까지만 기록**
    
    - **누적** : 최초 시점부터 해당 셀까지의 누적된 데이터를 보여줌
        - 3월 첫째 주체 획득한 사용자를 기준으로 한 달 동안의 데이터를 관찰한다고 할 때, +0주차부터 +4주 차까지 발생하는 데이터가 주 단위로 누적

<br>

- 세분화 : 코호트 분석을 보다 세분화해서 보고자 할 떄 사용
    - 세분화 기준은 변수 탭에서 가져온 측정기준을 활용

<br>

- 값 : 어떤 기준의 데이터를 출력할지를 정하는 것
    - 변수 탬에서 가져온 측정항목 중 데이터 호환성이 맞는 항목들만 활용할 수 있음
    - '측정항목 유형' 
        - 합계 : 모든 특성 행동을 모두 더함
        - 동질 집단 사용자당 : 비율로 데이터를 보기. 첫 주차 수치를 100%로 보았을 때, +1주 차, +2주 차 수치를 확인

<br>

#### 유입경로 탑색 분석
- 사용자의 전환 여정, 전환까지의 과정을 단계별로 분석하는 보고서
- 사용자가 각 단계에서 머무르는 시간, 전환으로 이어지는 경로, 중간에 이탈하는 지점을 확인할 수 있음
1. 이탈 지점 파악 : 어떤 단계에서 사용자가 떠나는지를 식별
2. 전환 과정 최적화
3. 고객 행동 이해 : 퍼널 단계를 다양하게 교차해 보면서, 고객이 구매 결정을 내리는 데 영향을 주는 요소를 이해

<br>

<img src='image\GTM\030106.png' width=700>

<br>

- '시각화'
    - 표준 유입경로  : 막대 그래프로 표현. 기준으로 하는 기간의 전체 데이터를 보여주기에 퍼널별로 어떤 구간에서 이탈이 많이 발생했는지 확인하는 데 용이
    - 인기 유입 경로 : 선 그래프로 표현되며, 각 퍼널 단계별로 일자별 데이터를 확인 가능
- '개방형 유입경로' 
    - OFF : 설정한 퍼널의 순서로만 이동한 사용자의 데이터만 출력
    - ON : 각 퍼널별로 중간에 유입자가 발생했다면, 이 사용자들의 데이터도 취합하여 보고


- **'단계'**
    - 최대 10개 단계 설정 가능

<img src='image\GTM\030107.png' width=500>

<br>

- '경과시간 표시' : 구간별 소요된 평균 시간을 확인할 수 있는 기능
    - **시간을 표시하는 숫자 자체에 의미를 두기보다는, 어떤 액션을 진행한 후에 이전 기간 대비 시간이 얼마나 증가/감소 했는지 파악하는 것이 바람직**


* '다음 작업' : 구간별로 그 다음 어떤 이벤트가 발생했는지를 확인할 수 잇는 기능
    - '이벤트 이름'으로만 선택 가능하며, 그래프 이에 마우스 포인터를 올리면 관련 이벤트 확인 가능

#### 세그먼트 중복 분석
- 마케터에게 타겟 사용자를 찾는 데 매우 유용한 도구
- 사용자가 서로 다른 세그먼트 간에 어떻게 겹치는지를 보여주어, 여러 세그먼트와 고객 행동이 어떻게 상호작용을 하는지 확인할 수 있도록 도움
- GA4 인터페이스에서 '세그먼트'를 생성하여야 함

<img src='image\GTM\030108.png' width=700>

<br>

- **새 세그먼트 생성**

<img src='image\GTM\030109.png' width=700>

- **'맞춤 세그먼트 만들기' : 특정 사용자, 세션 세그먼트, 이벤트를 기준으로 원하는 조건의 세그먼트를 생성**

    예) 어떤 사용자가 첫 세션에서 `view_item`이벤트를 완료하고, `view_promotion`이벤트를 발동. 두 번째 세션에서 `page_view`, `add_to_cart`, `purchase` 순으로 이벤트를 발생

    $\rightarrow$ `purchase`이벤트를 조건으로 이벤트를 세팅

    $\rightarrow$ **사용자 세그먼트는 '사용자 A', 세션 세그먼트는 '세션 2', 이벤트 세그먼트는 `purchase` 이벤트만 포함하게 됨**

    - 사용자 세그먼트를 클릭 $\rightarrow$ '새 조건 추가' 클릭 $\rightarrow$ 원 하는 조건 입력
    - '특정 시점에서 참' / '사용자의 마지막 상호작용에서 참'
        - 예) 사용자 한 명이 2개월 전 브라직, 현재는 다른 위치 $\rightarrow$ 조건이 '국가 = 브라질', 기간이 이전 3개월인 세그먼트를 추가

            $\rightarrow$ 사용자의 마지막 상호작용에서 참 : 세그먼트에 포함

    - '제외할 그룹 추가'
        - 예) PC를 통해 유입한 사람 중 특정 브라우저를 통해 유입한 사용자를 제외
    
    - '시퀀스' 
        - 예) A페이지에 접속 후 B페이지에 접속한 사용자들을 세그먼트 $\rightarrow$ '포함할 시퀀스 추가' (최대 10개)
    
    - '잠재고객 만들기' : GA4에서 생성한 세그먼트를 잠재고객으로 저장하여 잠재고객 보고서로 모니터링하거나 Google Ads와 연동하여 타겟팅으로 활용하게끔 도와주는 기능 (최대 540일 까지 보관 설정 가능)
        - 타겟팅으로 활용하려면 광고를 송출하고자 하는 Google Ads와 연동이 되어있어야 함
        - '잠재고객 트리거' : 설정한 조건의 잠재고객이 모집될 때마다, 설정한 이름으로 이벤트 데이터를 수집하고자 할 떄 활용

<br>

- **추천 세그먼트** : 사전 설정된 세그먼트 템플릿
    - '예상 검색어' : 예측 모델 기반
    - 구매 관련 예측 모델을 적용하기 위해서는 다음 3가지 조건의 충족 필요
        1. 구매자 및 앱 제거 사용자의 긍정적/부정적 예시 수가 최소 요건을 충족
            - 최근 28일 동안 7일간의 기간에 대해 1,000명 이상의 재사용자가 관련 예측 조건 (구매 또는 앱 제거)을 트리거했어야 하며, 1,000 명 이상의 재사용자는 트리거하지 않았어야 함
        2. 일정 기간 동안 모델 품질을 유지해야 요건을 충족
        3. 구매 가능성과 예측 수익 측정항목을 모두 사요하려면 `purchase` 또는 `in_app_purchase` 이벤트가 필요
            - `purchase`이벤트를 수집하려면, 해당 이벤트에 대한 `value` 및 `currency` 매개변수도 수집해야 함
        - 예측의 핵심은 구글 애널리틱스에 `추천 이벤트'를 통해 이해할 수 있는 데이터를 꾸준히 수집하는 것
        - 조건이 충족되는데 어느 정도의 시간이 필요한지는 알 수 없음

<br>

#### 경로 탐색 분석
- 페이지나 이벤트 경로를 바탕으로 사용자 행동의 흐름을 분석하는 방법
- 사용자의 흐름을 파악한다는 점에서 '유입경로 탐색 분석'과 유사한 목적을 가지지만, 경로 탐색 분석은 전체적인 사용자 흐름을 볼 수 있도록 도아줌

<img src='image\GTM\030110.png' width=700>

<br>

- 노드 (Node) : 사용자가 경로상에서 방문하는 개별적인 포인트나 단계
    - 사용자가 웹사이트나 앱 내에서 취한 각각의 행동이며, '페이지 뷰', '이벤트'와 같은 상호작용을 의미

<br>

- '고유한 노드만 보기' : 활성화 시, 사용자의 행동 경로에서 반복적으로 나타나는 노드를 한 번씩만 표시


<br>

#### 사용자 전체 기간
- 사용자가 처음 상호작용을 시작한 시점부터 현재까지와 전체 기간 사용자 행동과 가치를 종합적으로 분석하는 도구
- 사용자의 생애 가치 (LTV)를 이해하는 데 중요하며, 마케터가 장기적인 고객 가치를 평가하고, 전략을 수립하는 데 도움

<img src='image\GTM\030111.png' width=700>

- 사용자 전체 기간 보고서는 사용자가 서비스를 처음 방문한 이후부터, 플랫폼에서 완전히 이탈할 때 까지의 사용 패턴을 확인 가능

    이 보고서에서 중요한 것은 '사용자를 어떤 기준으로 구분할 것인지'와 '그 그룹별 비교'

예)  사용자를 아래와 같이 구분했다고 가정

| 그룹 | 평균 참여 시간 | 거래 횟수 | LTV |
| - | - | - | - |
| 유저 전체 평균 | 4분 | 1회 | 6$ |
| 그룹 1 | 7분 | 3회 | 15$ |
| 그룹 2 | 6분 | -회 | - |

<br>

- 그룹 1은 평균 사용자들보다 참여 시간, 거래 횟수, LTV가 모두 높음

    $\rightarrow$ 그룹 1을 VIP 고객군으로 명명

    $\rightarrow$ 마케터는 이 그룹에 보다 적극적인 CRM 마케팅 진행
- 그룹 2는 아직 거래를 진행하지는 않았지만, 사용자 전체 그룹보다 참여 시간이 높음

    $\rightarrow$ 아직 구매는 없지만, 구매 확률이 높은 잠재고객군으로 분류 가능

- 사용자 전체 기간 보고서에는 '예측'이라는 측정 항목이 제공

    - 예측 데이터를 활용하면 구매 확률이 높은 사용자에게 쿠폰을 제공하거나, 앱 제거 가능성이 높은 사용자에게 고객 관리 이메일을 전송하는 등의 CRM 마케팅에 도움이 되는 힌트를 얻을 수 있음
    
    (예측 항목을 제대로 활용하려면, '추천 이벤트'양식에 맞는 이벤트 명과, 매개변수 명으로 데이터가 수집되어야 함)

- '사용자 전체 기간 그룹' : 사용자의 첫 방문일 등 사용자 기준의 데이터를 확인 가능

    $\rightarrow$ 마케터는 고객의 장기적인 가치를 평가하고, 각 사용자 그룹벼로 맞춤형 전략을 수립할 수 있음
