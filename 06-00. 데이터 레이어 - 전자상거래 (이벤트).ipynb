### 전자상거래 이벤트의 정의
- 구글 애널리틱스는 우리 웹사이트나 어플리케이션에 방문한 유저들의 모든 활동(페이지뷰, 클릭 등)을 시간대별로 기록해둔 로그 분석 툴
- **최초의 Google Analytics는 쇼핑몰을 염두해두고 설계된 솔루션**
  
  $\rightarrow$ 그렇다보니 위 `[수익 창출]` 메뉴에 있는 다양한 리포트는 쇼핑몰 비즈니스를 하고있는 분들에게 도움이 되는 유용한 리포트들
  
  $\rightarrow$ 즉, 쇼핑몰이 아닌 비즈니스는 해당 리포트를 조회할 필요도 없고, 데이터가 들어오지도 않음
  
  $\rightarrow$ **하지만 쇼핑몰인 고객들은 유용하게 해당 리포트를 사용해야만 하는데, 해당 리포트를 사용하기 위해서는 전자상거래에 관련된 이벤트를 설치해야할 필요**
  
  $\rightarrow$ **구글 애널리틱스 내 전자상거래와 관련된 리포트를 보기위해 설치해주어야하는 이벤트들을 우리는 전자상거래 이벤트(E-commerce Event)**

## 전자상거래 이벤트

<br>

#### 전자상거래 이벤트는 모두 설치해야하는가?
- 전자상거래 이벤트를 잘 구현하기 위해서는 아래와 같은 14개의 GA4에서 제공해주는 전자상거래 이벤트가 존재. 
- 해당 14개를 모두 설치하실 필요는 없으며, 대한민국에서만 관행적으로 취사선택에 따라서 이벤트를 전송하는 경우 빈번 
  
  $\rightarrow$. 그렇기 때문에 오롯이 설치 여부는 구글 애널리틱스 담당자의 선택


<br>

### 01. `purchase` (필수)
- **발생 시점** : 구매가 완료되었을 때 발생
  - 한국 쇼핑몰의 경우 대표적으로 아래 2가지 경우

  1. 구매 완료 페이지가 로드 될때
  2. 네이버페이를 클릭했을 때
    - **네이버페이의 경우, 버튼을 클릭하면 사이트 밖을 나가버리게 되므로, 그 이후 시점에 대해서는 GA4가 데이터를 추적할 수 없음**
    
    $\rightarrow$ 현재의 기술로서는 네이버 페이 버튼을 클릭할때 `purchase` 이벤트를 발생시킴
    
    - 버튼을 클릭하고 구매하지 않는 유저들도 있을 수 있기 때문에, 과장집계가 발생 가능하며, 과장 집계가 발생되는것을 막기 위해서는 자체 서버를 통해 데이터를 전송하는 Measurement Protocol이란 방법론이 존재
    
    하지만 자체 트랜잭션 데이터를 받고 있는 서버나 DB가 구축된 환경에서만 가능하여, 대부분의 커머스들을 정확하게 구현하는데에 비용과 시간이 많이 발생
    
    $\rightarrow$ 그래서 관행적으로 간단하게나마 추이를 보는 용도로서 네이버페이 버튼을 클릭할때 `purchase` 이벤트를 전송하는것이 시장에서 일반적으로 통용되는 방법론

<br>

### 02. `add_to_cart` (필수)
- **발생 시점** : 장바구니 버튼을 클릭하여, 물건이 장바구니에 담긴 순간 발생
  - **장바구니 페이지가 로드되는 이벤트는 `view_cart` 이벤트이며, `add_to_cart`는  제품 상세페이지 내에서 물건을 장바구니에 담을때 데이터를 전송**

<br>

### 03. `begin_checkout` (필수)
- **발생 시점** : 결제를 시작하는 페이지가 로드되었을 때 발생

<br>

### 04. `view_item` (필수)
- **발생 시점** : 제품의 상세페이지를 조회할때(페이지 로드) 발생

<br>

### 05. `add_payment_info` (선택사항)
- **발생 시점** : 결제정보(카드 정보, 통장 정보 등)를 입력 완료하는 순간에 발생
- 대한민국의 경우, 네이버페이와 무통장입금 방법론이 일반적으로 활성화되어있는데에 반해, 글로벌에서는 신용카드 정보를 등록하고 구매를 자유롭게 하는 방식이 일반적
  
  $\rightarrow$ 해당 이벤트를 대한민국에서는 굳이 추가하지 않으며, `begin_checkout`(결제 시작)이라는 이벤트로 사용자의 행동흐름은 충분히 관측할 수 있기 때문.

<br>

### 06. `add_shipping_info` (선택사항)
- **발생 시점** : 배송지 정보(주소, 우편번호 등)를 입력 완료하는 순간에 발생
- 대한민국의 경우, 결제 시작 페이지에서 카드 정보와 배송지 정보를 한번에 적기 때문에, 굳이 배송지 정보 추가 이벤트를 일반적으로는 하지 않음
  
<br>

### 07. `add_to_wishlist` (선택사항)
- **발생 시점** : 제품 찜하기, 즐겨찾기, 위시리스트 추가 등이 이루어지는 순간에 발생
- 쇼핑몰마다 제품을 찜하는 방법은 다르지만, 의미론적으로 제품을 찜하는 어떤 고객의 행동이 정의될 수 있다면 해당 이벤트를 전송
  
  다만, 대한민국의 경우, 일반적으로 찜하기 기능을 대형 쇼핑몰이 아니라면 많이 사용하지 않는것 같아서, 잘 설치하지 않는 경향

<br>

### 08. `refund` (선택사항)
- **발생 시점** : 제품 구매 취소를 하거나 환불 신청을 하는 경우 발생
  - 전화로 환불을 했을 경우에는 이벤트를 잡을 수가 없음. $\leftarrow$ 사이트에서 발생하는 액션들을 중심으로만 이벤트를 발생시킬 수 있기 때문
- 사이트 UI에 구매 취소나 환불 신청 완료페이지가 있거나 등의 고객의 행동을 잡을 수 있는 트리거 시점이 있다면 해당 이벤트를 전송할 수 있음
  
  $\rightarrow$ 환불 이벤트를 적용하느냐 아니냐에 따라서 리포트 내 측정 항목들이 달라질 수 있음

<br>

### 09. `remove_from_cart` (선택사항)
- **발생 시점** : 장바구니에 담긴 물품을 장바구니에서 빼는 순간(장바구니 담기 취소) 발생
- 장바구니 취소 이벤트 역시 데이터를 적재한다고 해서 크게 인사이트가 있는 내용을 담기 어렵기 때문에, 한국에서는 관행적으로 많이 설치하지 않음
  
<br>

### 10. `select_item` (선택사항)
- **발생 시점** : 어떤 목록에 있는 제품을 클릭하는 순간 발생
    
    (여기서 ‘목록(list)’이란 여러 항목들이 한 영역에 나열되어 있는 형태를 의미)
    
    아래 이미지와 같이 특정 영역에 있는 진열된 상품이 클릭 되었을 때 추적

    <img src ='https://lh7-us.googleusercontent.com/CNfQdmMySdnvPs5Vrqpho-nk6E_Tq3nI5WNOwEnsrefllhbks4dxTj1DaqH0BjUaO_1Waj6EIx3j1_y3jJyJ9j_5iJJUPpWBLA_CfP-vxrcwq4c6kk-kwsJT4MzDYF168SbiMJrXkbH3EMVIiGAOL1I' width='500'>

    - 사이트에 진열된 제품 영역들이 정말 많아도, 모든 영역들을 정의하여 예외없이 데이터를 전송하도록 해야함.

    모든 영역들을 조사한 뒤, 각 영역별로 제품이 클릭되었을 때 이벤트를 발생시키도록 설정

    $\rightarrow$ `select_item`은 `purchase`, `add_to_cart와` 같은 일반적인 전자상거래 이벤트보다 훨씬 더 많은 설계 및 개발 공수가 발생


<br>

### 11. `select_promotion` (선택사항)
- **발생 시점** : 특정 형태의 목록에 있는 프로모션을 클릭할 때 발생
- 해당 이벤트는 사이트 내에 있는 프로모션(행사,기획전 등)과 관련된 영역/버튼/배너 등을 클릭하는 경우 이벤트를 발생
  
  아래 이미지를 보면, 프로모션 영역이 명확하게 표현되어있습니다. 아래 이미지에 표시된 프로모션 영역을 클릭하는 경우 이벤트를 발생

  <img src='https://lh7-us.googleusercontent.com/3Fm7F7AMCn27o8Em5jMbkQ_JCQE2Fs1UZCm09woyfV9RRFQ5DOygz0WhrlTmWunnbCrvJriyHdP6rT2Tt3KrHn0IVOhr9clojaSyPQBtwRRlkU9mgRd_Jr3zhzejph50UJrzZQR__YKqwaS8F72X-tA' width='500'>

- 프로모션 이벤트 역시 사이트에 정말 다양하게 있을 수 있으며, 해당 영역 역시 `select_item` 이벤트와 마찬가지로 많은 공수가 발생할 수 있음
  
  $\rightarrow$ 무엇을 프로모션 영역이라고 지정해야하는가?

  $\rightarrow$ **결국 구매를 촉진시키기 위해 고객에게 혜택을 제공하고 있는 모든 영역은 ‘프로모션(promotion)’**

  (그럼에도 불구하고 애매한 경우들이 많이 발생)
  
  $\rightarrow$ 프로모션 영역을 어디까지만 지정할것인가에 대한 사람의 의사결정이 필요

  <img src ='https://lh7-us.googleusercontent.com/cHPpUyzIbIt86ilsiX2ZBiKz5afMU8I7uSFofCnYup6iiM3bR7iXstv4iLINrNUjTzskdtjzwVd8EwiKqUOme7SmWZdHVO8jdcmKZvRXvYTvmnliJRKYaZZ6Jz1ze0oHVcH4InbGPktZhuaNrCCeFWQ' width='300'>

  위에 있는 영역(”모든 서비스는 0원”)은 확실히 고객에게 어필하는 혜택이기 때문에 프로모션 영역이 확실
  
  다만 아래있는 영역(”골프존 사천 205,000원~”)은 상품이 진열된 영역인것 같기도하고, 최저가를 제시하다보니 프로모션 영역인것 같기도 함

  $\rightarrow$ 이 경우에는 개인의 취사선택에 따라 프로모션 영역으로 잡을것인지 진열된 상품 영역으로 잡을 것인지를 결정
  
  (다만, 하단에 있는 영역은 클릭했을 때, 그 상품의 상세페이지로 이동되기 때문에 굳이 프로모션 영역이라기보다는 아이템 진열 영역으로 잡는것이 더 합리적)

<br>

### 12. `view_cart` (선택사항)
- **발생 시점** : 장바구니 페이지가 로드되는 시점에 발생
- 보통 대한민국에서는 장바구니 페이지를 거치지 않는 네이버페이 결제 방식이 많고, 장바구니 담기 버튼을 클릭할때, `add_to_cart` 이벤트를 발생시키기 때문에 굳이 디테일하게 `view_cart` 이벤트는 많이 설치하지 않는 경향

<br>

### 13. `view_item_list` (선택사항)
- **발생 시점** : 제품이 있는 목록에 제품이 화면상에 노출되었을때 발생

    <img src='https://lh7-us.googleusercontent.com/0U7k-jdZFXxJh9gh1mnBmZ2tFiSVnUztGEF6GyReWN-zUd3vFw9ZuOAW6U-qSsWhfkDsHMFdHtuHd9u9f5ar0nepeUf19Zo-5WgrkO17oIjoCDJjXhtZG9yYWrRZbkikWxKdHe_BovohyOKBRgyDUc8' width='500'>

- 위 이미지와 같이 쇼핑몰을 방문한 유저의 브라우저 화면에 제품이 나타났을 때 이벤트를 발생
    
    (총 5개의 제품이 화면상에 나타났기 때문에, 5번의 같은 이벤트가 발생할 것)

- **특정 영역이 화면에서 몇% 노출된 시점에서 몇초가 지나야 ‘노출’이 되었다고 판단할 것인지에 대한 정의가 필요** 
  
  하지만 모든 영역마다 기준이 있어보았자 데이터를 통합적으로 보는데 다른 기준을 가지고 보게 된다면 인사이트를 추출하기가 매우 어려움
- 위 이미지에 보이는 각각의 제품 영역은 하나의 화면에 여러개가 담길 정도로 크기가 작기 때문에(모바일도 포함), 
  
  100% 노출된 시점부터 1초가 지났을 때 `view_itme_list` 라는 이벤트를 전송
  
  해당 노출의 정도와 시간도 관행적으로 1초를 잡는것이지, 필요에 따라 다른 정의와 기준을 내려도 무방

<br>

### 14. `view_promotion` (선택사항)
- **발생 시점** : 프로모션 영역에 특정 프로모션 영역이 화면상에 노출되었을때 발생

    <img src='https://lh7-us.googleusercontent.com/FaehUFj7QXAUrNK1nTIp-nwEUXJOMHOB_L2TtBwCQAM8IeXkmBFaQy8XpAEwvey5eNR3PFKw18JS-RIM4ey1oST_G6K5WAolGToNfw9r6b6ADDYFjhd1sxrTF2JG-ViZh218wkqYwdq9ZLwQInqh-70' width='500'>

- 위 이미지와 같이 쇼핑몰을 방문한 유저의 브라우저 화면에 프로모션 배너가 나타났을 때 이벤트를 발생
    
    (총 4개의 배너가 화면상에 나타났기 때문에, 4번의 같은 이벤트가 발생할 것)
    
- **특정 영역이 화면에서 몇% 노출된 시점에서 몇초가 지나야 ‘노출’이 되었다고 판단할 것인지에 대한 정의가 필요** 
  
  하지만 모든 영역마다 기준이 있어보았자 데이터를 통합적으로 보는데 다른 기준을 가지고 보게 된다면 인사이트를 추출하기가 매우 어려움
- 위 이미지에 보이는 각각의 프로모션 영역은 하나의 화면에 여러개가 담길 정도로 크기가 작기 때문에(모바일도 포함), 
  
  100% 노출된 시점부터 1초가 지났을 때 `view_promotion ` 라는 이벤트를 전송
  
  해당 노출의 정도와 시간도 관행적으로 1초를 잡는것이지, 필요에 따라 다른 정의와 기준을 내려도 무방

## 전자상거래 이벤트에는 매개변수가 고정

<br>

#### Step 1 : 애널리틱스 `[수익 창출]` 카테고리에서 `[전자상거래 구매]` 리포트로 이동

#### Step 2 : 대표적인 측정항목 3개

<img src='https://lh7-us.googleusercontent.com/tMz_SChAX5MbMN5i1xnLU6qzN1RS_MwLgYzhU4yXj04p7fz_JuP7y1gJKF48Mhc4j_i0hN1kJ9kmDFt3Op1PmGZP6_MrZv5yraTVFbsRVjyB4AzkxjJKa5LXe8oUhxwx7tAsPYxCheokWqHjaKNF62g' width='600'>

- **구매한 상품** : `purchase` 이벤트가 발생해야만 데이터가 쌓임
- **장바구니에 추가된 상품** : `add_to_cart` 이벤트가 제품별로 발생했을 때마다 데이터가 쌓임
- **조회된 상품** : `view_item` 이라는 이벤트가 각 제품별 상세페이지가 로드될때마다 발생되어야만 데이터를 볼 수 있음
  

    $\rightarrow$ 전자상거래 이벤트 14개 중 몇가지를 설치하면 GA4에서 제공해주는 측정 항목이랑 ‘자동 연동’이 되면서 리포트에 숫자를 채워줌. 
    
- 위 이미지를 자세히 보면, `[항목 이름]` 이라는 측정기준별로 데이터를 관측
  - 항목 이름은 상품의 이름을 의미
  - 단순히 `add_to_cart` 이벤트만 보내게 되었다면 상품의 장바구니 담기수는 나오지만 상품 이름별로 관측하기에는 어려움 $\rightarrow$ `view_item` 이벤트만 보내는 것이 아닌, 해당 이벤트에 ‘상품 이름’을 의미하는 데이터를 매개변수로 보내주어야 함
  
  $\rightarrow$ `view_item` 이라는 이벤트에 `item_name` 이라는 매개변수가 필요
  
  **일반적으로 만들게되는 맞춤 이벤트와 달리, 전자상거래 이벤트의 매개변수는 명확하게 고정**

## 전자상거래 이벤트를 이해하기 위한 프로그래밍 

- 구매라는 이벤트가 발생했을 때, 이벤트와 매개변수별로 데이터를 보내면 된다고는 말했는데, items 라는 새로운 형태로 데이터가 보내지고 있는것을 확인

    $\rightarrow$ **제품을 동시에 여러개를 장바구니에 담을 수 있기 때문**

    <img src='https://lh7-us.googleusercontent.com/NXsSpVZ4qKJOAR3xgBYp_g1-aHr5cLKsQVEO7rJglJ_SFwBpT7lv4pmB0x-ZiT3ECfUEYXgdZogsrMUtu4K1q1jxxbyb7lymM5Z_onDa-6A12B-LQlDzLmMhufPC00yXsuvu9y1W3ulK27OinoWZza8' width='500'>

    - 1개 이상 제품을 구매했을 때에는 정상적으로 코드를 보내고 있으나,  
    
      2개 이상의 제품을 장바구니에 보내주기 위해서는 A제품에 대한 정보 덩어리와 B제품에 대한 정보 덩어리가 각각 보내져야 함
      
      $\rightarrow$ 나중에 GA4에서 각각의 제품별로 추가 정보를 보아야할 필요가 있기 때문

- 제품 덩어리가 하나씩 추가 될때마다, `“{}”`의 특수기호로 제품의 덩어리 세트를 만들어서 GA4로 데이터를 전송
- 자바스크립트에서는 각각의 데이터의 형태들이 필요나 목적, 규칙에 따라 기본적으로 나뉘어져있음

## 이벤트 수준의 변수와 아이템 수준의 변수

<br>

### 1) 이벤트 수준의 변수(Event-Level Variables)
- 이벤트에 추가 정보를 붙이는 것을 이벤트 수준의 변수
- **아래 코드처럼 구매 이벤트가 발생했을 때 거래 ID, 총 수익, 화폐단위 등의 데이터를 보내는것은 제품 개별적으로 보내는 것이 아니라, 구매 이벤트에 들어가야할 매개변수이기 때문**

<br>

### 2) 아이템 수준의 변수(Item-Level Variables)
- 각 개별 상품에 따른 추가 정보들을 붙이는 것을 아이템 수준의 변수
- 아래 코드처럼 구매 이벤트가 발생했을 때, 각 인덱스별로 `item_id`, `item_name` 등의 변수들은 이벤트에 딸려있는것이 아닌 `items`라는 개별 상품에 대한 정보

## `Items` 객체 데이터

<img src='https://lh7-us.googleusercontent.com/E5C1SNdx_PTCjB1X5ZUlc_bZFfRmQVFmsT1FMONW4P7u79vWreJ9u_VQP-e_uOZsK4XdsQRjvxI2RBauIuLa1AEyRMjrPMFMM69V7ZLt8-xZHRFGP0KXqhRydtBsTIXK1kvid2ruCKoA9ypVLp6BEk0' width='500'>

- 전자상거래 구매 리포트를 보면, 제품별 구매수와 수익이 표시 (여기서 구매한 상품은 제품별 구매 수를 의미) 
    
- 구매한 상품이 데이터에 잡히는 이유 : `Purchase` 이벤트를 보내주었기 때문
- 항목 이름에 제품명이 데이터에 잡히는 이유 : `Purchase` 이벤트를 전송할때 `items` 객체 형태로 데이터를 보내고, `items` 객체 안에 있는 `item_name`이라는 Key값의 Value 값이 GA4의 ‘항목이름’이라는 측정기준에 반환
- 제품 수익이 데이터에 잡히는 이유 : `Purchase` 이벤트를 전송할때 `items` 객체 형태로 데이터를 보내고, `items` 객체 안에 있는 `price`라는 Key값의 Value 값과, `quantity`라는 Key값의 Value 값이 서로 곱해져서, GA4의 ‘상품 수익’이라는 측정항목으로 반환

    위 예시와 같이 `items` 배열안에 들어가는 제품 객체의 정보들을 각각 GA4에서 특정한 측정 기준과 측정 항목을 채워주고 최종적으로 리포트를 만들어주기 위해 역할

<br>

- GA4에서의 리포트는 크게 2가지 종류로 분류
  1. GA4가 미리 준비한 템플릿 형태의 리포트인 표준 리포트(Standard Report)
  2. 내 입맛에 맞게 리포트를 구성하는 맞춤 리포트(Custom Report)
- GA4에서는 표준 리포트에서 쓰이는 측정 기준과 측정 항목의 이름, 맞춤 리포트에서 쓰이는 측정 기준과 측정 항목의 이름이 일부 다르기도 함 (많은 사용자들이 불편함을 겪는 부분) 
  
  심지어 맞춤 리포트에는 지원되는 측정 기준이 표준 리포트에는 없는게 있을 수 있음

## `Items` 객체 데이터를 각 이벤트마다 똑같이 일치

- 두번째 이미지의 경우, `purchase` 이벤트에는 일관되게 청바지가 `item_name`의 `value` 값으로 들어가는게 아니라, 원피스가 들어가 있으니, 원피스에서만 구매가 발생한것으로 기록
- 실제 데이터에서는 원피스는 `view_item`과 `add_to_cart`에는 없으니 조회된 상품도 0이고, 장바구니에 추가된 상품도 0
