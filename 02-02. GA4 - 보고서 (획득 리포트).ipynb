### 1. Paid Search
- 유료 검색 광고 (대표적으로 네이버, 구글, 다음, 카카오 다음 검색 광고 등)
  - 검색을 통해 유입되긴 하지만, ‘Paid’는 돈을 지불한 광고를 의미하므로, 검색 광고를 통해 유입된 것
- 대표적인 예로 네이버의 파워링크 광고
  - 네이버에서 “강남역 맛집”을 검색해보면, “장인닭갈비 강남점”이라는 광고가 표시 $\rightarrow$ 파란색으로 표시된 파워링크 영역은 모두 광고를 집행한 업체들이 돈을 지불한 결과

### 12. Email
- Email을 통해 방문한 경우는 Email 채널로 분류

<br>

### 13. SMS

- 고객들에게 문자 메시지를 발송하는 방식. 문자 메시지를 통해 방문한 경우는 SMS 채널로 인식

<br>

### 14. Mobile Push Notifications

- GA가 UA에서 변경되면서 도입된 기능
- 웹사이트만의 분석이 아닌, 애플리케이션 내에서도 고객들에게 푸시 알림을 전송하는 방식을 포함
- 푸시 알림을 통해 방문한 고객들은 Mobile Push Notifications 채널로 분류

<br>

### 15. Referral

- Referral은 한국어로 ‘추천’이라는 의미를 가지며, 다른 도메인에서 우리 웹사이트를 언급하고 링크를 통해 방문한 경우 해당 채널로 분류
- 예)
  - 개인 블로그에서 어떠한 상품에 대한 글을 작성하고 해당 상품의 웹사이트 링크를 공유했을 때, 그 링크를 통해 방문하는 사용자들은 Referral 유저로 인식

<br>

### 16. Affiliates

- 해당 채널은 우리의 계열사나 파트너사를 통한 방문을 의미
 
  따라서 Affiliates는 계열사나 파트너사를 통한 Referral 방문을 구분하여 파악하기 위한 채널

<br>

### 17. Direct
- **Direct의 정의는 Referral URL이 없거나 유실된 경우에 해당 사용자의 접속을 의미**
  - 주소창을 통한 직접 접속, 즐겨찾기, 바탕 화면 바로 가기 등에서는 이전의 URL 정보가 존재하지 않으며, 
    
    또한, 친구들 간에 카카오톡 등의 애플리케이션을 통해 공유된 링크로 접속할 때도 Referral URL이 존재하지 않기 때문에 Direct로 분류

- 다양한 상황에서 Referral URL이 유실될 수 있음
  - 예) 보안이 적용된 https 사이트에서 보안이 적용되지 않은 http 사이트로 이동할 때 Referral URL이 유실되는 경우

## 02-06. 획득 리포트 활용

<img src='https://lh7-us.googleusercontent.com/buDseozuL4SbI-X0at5A11oI__5_AKkV4_-yVwW7hLIN646FF0Q9SHAP0GEpbcQKgDOnGmHrOfOHF_M5wwmgCi7xv48otakMUe48UZZe5ZSlc20yqyBhDEljk3rUlfdvZIAHL7tc5A51G2uX8X6QB4s' width='600'>

- 만약 Direct의 트래픽이 지나치게 많다면, 다양한 경로로부터 오는 사용자들이 많다는 것을 의미
  -  그러나 이러한 사용자들이 어디서 오는지 정확히 알 수 없다는 것은, 다른 채널들에 대한 통제력을 잃어버릴 수 있다는 점을 암시
  
  $\rightarrow$ **그래서 Direct 유입자가 단순히 높아지느냐 낮아지느냐에 문제가 아니라, 전체 대비 차지하는 비중이 얼마나 되는가, 그 비중이 날짜별로 줄어들고 있느냐 높아지고 있느냐를 관측할 필요**

- **Direct의 트래픽을 완전히 없애는 것은 어렵지만 우리는 이런 Direct 트래픽을 점진적으로 줄이기 위한 노력이 필요 $\rightarrow$ UTM 매개변수**

<br>

- 대다수의 구매가 Direct로부터 발생한다면, 이로 인한 데이터의 제한성이 생길 수 있음
- 채널 리포트에서는 채널 간 전환율을 단순히 비교하는 것은 의미가 없음
  - 예) Organic Search와 Display 유저 간의 전환율을 비교한다면, Organic Search 유저의 전환율이 높은 것은 당연
  
  $\rightarrow$ Organic Search의 유저는 구매 의도를 가지고 웹사이트를 방문하는 반면, Display는 구매 의도가 없이 배너를 클릭하여 웹사이트를 방문하기 때문

  $\rightarrow$ **따라서, 채널 리포트에서는 시간의 맥락, 예를 들면 전일 대비, 전주 대비 등을 고려하여 데이터를 분석**

## 02-07. 사용자 획득 리포트와 트래픽 획득 리포트의 차이
- **트래픽 획득리포트를 보게 될 경우, 사용자 획득 리포트와 달리, 세션의 기본 채널 그룹으로 측정 기준이 변경**
  
  **원래는 ‘신규 사용자 기본 채널 그룹’이었지만, 이제는 ‘세션 기본 채널 그룹’으로 나타남**

  <img src='https://lh7-us.googleusercontent.com/ntfXzQQ1gvktIdiBL38KjJFkSRiZl9bJ1bIRL_wOSrc4qobyjXLWD6ghh0t1CGm4rTNSIJsd1lPlyvkwh5Fu8CgiHRGu-aYf4NEluyBUWB5Vv8_GGMerLbSd9-xFKs2zDMOcyPrl8KLAvqiMXkPgEiY' width='500'>

  <img src='https://lh7-us.googleusercontent.com/aencVQ71yY7p5nlzfRzbmoGJ7hCaotzD5Xuq_oSOOqwqPR8kyDvfB7o21QX2p_miP4QMOUIlgXVQrYyEIsGQ6YNZTddmIJffhZ7xNiwUuzS56IdTZPyoEuGN4xapJf2ArrnMtKNczwR52ONcXz0DnVM' width='500'>

<br>

#### 예) A 사용자가 1월 1일에 네이버로 검색하여 웹사이트를 방문한 후, 1월 5일에 페이스북 광고를 통해 다시 방문 :
  
$\rightarrow$ 사용자는 한 명

$\rightarrow$ 세션 수는 두 개

$\rightarrow$ **세션 기본 채널 그룹은 사용자의 세션을 기준으로 유입 경로를 계속해서 리포팅**

<br>

- ‘사용자 획득’은 무엇을 의미? 
  
  $\rightarrow$ 이 리포트는 신규 사용자만을 대상 $\rightarrow$ 따라서 A가 처음 네이버에서 검색하여 웹사이트를 방문한 경우, ‘Organic Search’에만 신규 사용자로 기록
  
* GA4는 웹 분석뿐만 아니라 앱 분석도 포함
  * 앱을 사용하는 유저는 앱을 다운로드 한 후, 다시 방문할 때 대부분 ‘Direct’ $\rightarrow$ 이러한 이유로, 앱 마케터는 신규 사용자가 어느 매체를 통해 앱을 다운로드했는지에 초점을 맞추어야 함.

<br>

* **GA4는 웹과 앱 분석을 동시에 수행하는 툴이기 때문에, ‘사용자 획득’ 리포트와 ‘트래픽 획득’ 리포트를 모두 제공**
  
  $\rightarrow$ **웹 마케터는 주로 ‘트래픽 획득’ 리포트를, 앱 마케터는 ‘사용자 획득’ 리포트를 주로 참조**
