#### GA4의 메뉴는 총 5가지

<img src='https://lh7-us.googleusercontent.com/PH7dYhjtXJoOBBbg7f86T3JV1CyogjzaHEsjdAsUkbKAawTx9uXmnAIdRZ9Umn2nMODnNCpudd2dW5HsWh5zSrRNkixNsP8syQb5fkB-aNyVVOZMX04O46AdkNLlB9c33_LCbZnEplnlvIru7Pzk7SA' width='600'>


1. **홈** : 홈 화면은 사이트의 전반적인 지표들이 대시보드의 형태로 보고
- 전체 사용자 수, 새 사용자 수와 같은 거시적인 측면에서의 지표
- 홈 화면의 대시보드는 추후 내가 원하는대로 대시보드를 만들수도 있지만, 상세 분석을 하기에는 용이하지 않음

<br>

2. **보고서** : 구글 애널리틱스에서 기본적인 템플릿의 형태로 제공하는 리포트들을 ‘표준 리포트(Standard Report)’
- 표준 리포트들이 모여있는 영역이 ‘보고서’메뉴

<br>

3. **탐색** : 내가 원하는대로 보고싶은 항목들을 조합하여 나만의 맞춤 리포트를 만들 수 있는 영역이 탐색 리포트

<br>

4. **광고** 
- 고객들은 다양한 경로들을 통해 우리 웹사이트에 방문하며, 웹사이트에 방문한 고객들은 구매를 발생 
- 원하는 행위가 어떠한 경로로 발생했는가를 보여주는 지표
- 특히, 추후 구글 광고를 연동하면, 광고의 비용 데이터를 가져와서, 광고 비용 대비 매출이 얼마나 발생했는지에 대한 ROI 지표도 제공

<br>

5. **관리** : 구글 애널리틱스의 관리자 화면으로 이동할 수 있는 메뉴
- 구글 애널리틱스의 이벤트 설정, 디버그 뷰, 맞춤 측정 기준 설정 등 우리가 원하는대로 데이터를 조작하고, 세팅하고, 제어할 수 있는 전반의 기능들이 있는 관리자 메뉴로 이동

<br>

#### `[보고서 개요]` 
- 오른쪽 상단에는 날짜 필터. 웹사이트의 데이터를 관측하는 날짜를 조정 (항상 오늘 날짜로부터 지난 28일이 기본값으로 설정)

#### `[실시간 개요]`

- 실시간 리포트에서는 데이터가 쌓이지 않으며, 최근 30분동안의 행위만 보고하고, 30분이 지나면 휘발

#### `표준 리포트`는 항상 `[개요]`와 `[세부정보]`로 구분

<img src='https://lh7-us.googleusercontent.com/FXXJ_CWRbtUhsu0giL5EnFxUAY_HakcoIHmeIhLCsTav3m1K4eG6wHux79h5hubaGDje-pioENP_w7L6p6WG-CV6E7xkkFPXXH43FYCKfvwH_O025Pn2GeS2nhiSPqB7twGmE9pGPifMHrnZ3MTpWoA' width='600'>

- 구글 애널리틱스 `[보고서]` 메뉴를 클릭 $\rightarrow$ `[보고서]` 메뉴에 $\rightarrow$ `[보고서 개요]` 를 시작으로 모든 표준 리포트들이 각각의 용도에 맞게 분류
  - `[사용자]` 카테고리에 있는 `[인구통계]` 카테고리를 클릭 시, `[개요]`와 `[인구통계 세부정보]` 메뉴가 표시

1. **`[개요]`** : 개괄적인 지표들을 보여주기 때문에, 실무적으로는 많은 인사이트를 얻을 수 없음

2. **`[세부정보]`** : 각 표준 리포트들의 자세한 데이터들을 테이블 차트로 표현해주기 때문에, 유용한 정보들과 인사이트를 얻을 수 있음

## 02-02. 측정기준(Dimension)과 측정항목(Metrics)
- `[인구통계 세부정보]` 리포트

<img src='https://lh7-us.googleusercontent.com/meCfwlysL-CPFoUi5jvye2OoNjk_8xHX2AtmjVk8DtsE2bSODydYkWXdtwzZElp0SyB-P43cTs-sqIDn7U-VAqTY1D1VCCfz6MAX5QY1xFUGx4Q4zi5WpzSwTgr4CkJnTDnVrtV4rIm0nRMmGaTPVS0' width='600'>


1. **시계열차트** : 시간의 흐름에 따른 지표의 변화를 보여주는 꺾은선 차트를 의미

2. **가로 막대 그래프** : 국가별로 사용자수를 비교하기 위한 용도로서, 가로 막대 그래프

3. **테이블 차트** : 실제 데이터들을 자세하게 관측할 수 있도록 행과 열로 만들어진 차트를 ‘테이블 차트’

- **하나의 숫자로 표현되는 각각의 항목들을 ‘측정항목(Metrics)’**

## 02-03. GA4의 세션(Session)과 참여(Engagement)

<br>

### 세션
- GA4에서 세션이란, ‘특정 사용자의 연속된 서비스 사용을 나타내는 단위’

<br>

### 참여
- ‘참여’란 지금 아래 세 가지 조건 중에 한 가지라도 만족한 세션이라면 참여한 세션이라고 판단

1. 웹사이트에 머문 지 10초가 지난 경우
2. 페이지 조회 수가 2회 이상인 경우(페이지 뷰 수 2회 이상)
3. 특정 중요한 이벤트(전환)가 1회 이상 발생했을 경우

<br>

### GA4 이벤트와 전환
- **이벤트(Event)** : 우리 웹사이트나 어플리케이션에서 발생하는 특별한 사건
- **주요 이벤트(Key Event)** : 우리가 추적하고 있는 이벤트들 중 특히나 중요한 이벤트

## 02-04. 기술 세부정보 리포트
- 기술 세부정보 리포트에 접속하면, 가장 먼저 눈에 띄는 것은 브라우저를 기준으로 한 측정 기준
- 이를 통해 웹사이트에 방문하는 고객들이 어떤 브라우저를 사용하는지 확인할 수 있으며, 이 정보는 UX 팀에게 특정 브라우저에서 발생하는 문제점을 파악하는 데 도움

- 이 모든 측정 기준들은 웹사이트 방문자들이 어떤 기기나 환경에서 접속했는지를 나타내며, 
  
  기술 세부정보 리포트는 이런 디바이스 환경에 초점을 맞춘 것이고, 인구 통계 세부정보 리포트는 방문자의 인구통계적 특성을 중점
