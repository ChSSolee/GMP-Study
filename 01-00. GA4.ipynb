# 01. 분석툴 활용의 필요성
- 플랫폼 및 플랫폼과 연계된 앱이나 웹에서 사용자가 활동을 하게 되면, 그 활동이 데이터화 되어, 
  
  해당 플랫폼을 소유한 IT 기업 서버에 저장

  $\rightarrow$ 그 데이터를  보유하게 된 기업들은 이를 활용하여 다양한 비즈니스 실행 $\rightarrow$ **광고 상품 개발**

- 오늘날 일반적으로 한 사람에게 노출되는 광고(메시지)의 수는 3,000 ~ 10,000개 사이,
  
  $\rightarrow$ 이러한 시대에서 예산은 제한이 있는데, 활용할 수 있는 광고의 영역은 너무 많음

  $\rightarrow$ 구글 애널리틱스와 같은 분석 툴을 활용하면, 매체 선정에 도움이 되는 정보를 얻을 수 있으며,
    
    사이트에 유입하였거나, 구매하는 사용자들에 대한 정보를 분석 툴의 기준으로 어느 정도 그 특징을 파악할 수 있음

    $\rightarrow$ **즉, 분석 툴을 활용하면 광고의 영역에서도 어떤 매체 혹은 어떤 메시지를 통해 들어온 사용자가 더 결이 맞는지 판단할 수 있고, 내가 운영하는 웹 사이트나 앱에 유입되는 사용자들의 행동 정보를 얻을 수 있음**

<br>

- 현재 많은 종류의 분석툴이 존재 (예: Amplitude, Mixpanel 등) 하지만, 그 중 Google Analytics를 사용하는 3가지 이유
1. **무료** 
- 유료 버전 (360)도 존재 : https://support.google.com/analytics/answer/11202874?hl=ko
  
<br>

2. **풍부한 정보**
- 무료 버전에서도 강력한 분석 기능 $\rightarrow$ 많은 참고 자료

<br>

3. **타 구글 마케팅 툴과의 높은 연동성**
- 구글은 구글 애널리틱스 외에도 구글 태그 매니저, 구글 애즈, 빅쿼리 등 다양한 툴 을 보유

## 02. 구글 애널리틱스의 구조
- **GA4는 계정, 속성, 데이터 스트림 3단계 구조로 이루어짐**
  - 계정 안에 속성이 포함되어 있고, 속성 안에 데이터 스트림이 다시 포함된 형태
    
<img src='image/GTM/0103.png' width='700'>
  
<br>

### 02-01. 계정
- 계정은 GA4에 접근하기 위한 최상위 단위
- 각 사용자는 최대 100개의 GA4계정을 소유할 수 있으며, 계정별로 분석하고자 하는 프로젝트를 관리할 수 있음
- 계정에서는 운영 관리와 관련한 기능을 제공
- GA4에 접근하는 사용자의 권한 및 액세스 관리에 대한 설정 가능

<br>

#### 계정 단계에서 설정 가능한 활용 사례
- A라는 마케팅 에이전시가 여러 클라이언트의 웹사이트와 앱을 관리하는 상환
  
  $\rightarrow$ 에이전시는 다양한 클라이언트에 대응하기 위해, 클라이언트 단위로 별도의 GA4 계정을 직접 생성 또는 권한 부여를 통해 접근 가능

  $\rightarrow$ 클라이언트 데이터 관리가 용이해지고, 마우스 클릭 몇 번으로 클라이언트의 GA4 데이터로 빠른 이동 가능

<br>

### 02-02. 속성
- 속성은 특정 웺사이트나 앱의 데이터를 수집, 관리, 분석하는 역할
- 각 속성에는 데이터 스트림이라는 데이터 수집을 위한 스크립트를 포함 $\rightarrow$ 사용자와의 상호작용 데이터를 추적
- 속성에서는 사이트 방문, 앱 사용, 사용자 행동 등 다양한 데이터를 수집하고, 이를 바탕으로 분석을 수행
- 데이터 수집 외에 잠재고객, 맞춤 정의 채널 그룹 설정 등이 가능

<br>

#### 계정과 속성의 구조화
- 한 기업이 여러 브랜드의 웹 사이트를 운영한다고 가정

   $\rightarrow$ 하나의 GA4 계정을 보유하면서, 각 브랜드 웹사이트들 별로 속성을 분리

   $\rightarrow$ **하나의 GA4 계정에서 속성 단위로, 각각의 브랜드 사이트 데이터를 관리하고 분석 가능**

<br>

#### 데이터 스트림
- 데이터 스트림 (Data Stream)은 속성 내에서 웹, iOS, Android와 같은 특정 플랫폼으로부터 데이터를 수집하는 구체적인 방식
- 각 속성은 하나 이상의 데이터 스트림을 가질 수 있으며, 이를 통해 다양한 출처에서 오는 데이터를 통합하여 분석 가능

<br>

#### 웹 & 앱 (크로스 도메인)
- 한 회사가 어떤 브랜드를 보유하고 있는데, 그 브랜드가 모바일 앱과 웹 사이트를 동시에 운영하는 경우
  
  $\rightarrow$ 브랜드 담당자는 웹과 앱 각각을 위한 데이터 스트림을 별도로 설정하여, 브랜드와 관련한 플랫폼에 유입하는 사용자들의 통합 분석 가능

<br>

## 03. 구글 애널리틱스4의 특징
- 2020년 10월 GA4가 본격적으로 런칭

<br>

### 이벤트 기반 추적 방식
- GA4의 가장 큰 특징은 데이터를 수집하는 방법이 완전히 변경 (UA (GA3)은 세션을 기반으로 데이터를 추적)
  
  $\rightarrow$ 사용자와의 상호작용을 더 유연하고 세밀하게 추적 가능

- **이벤트 기반 데이터 수집 모델은 페이지 뷰, 버튼 클릭, 동영상 재생 등 거의 모든 사용자 상호작용을 '벤트' 라는 작은 단위로 수집**

<br>

### 유연해진 데이터 변경
- '필터', '기여 모델'에 대해 자유롭게 데이터가 변경이 가능해졌으며, 과거 데이터를 현재 버전의 설정으로 변경되는 소급 적용까지 가능
  
  (과거 UA에서는 필터를 한 번 설정하면 필터로 인해, 수집되지 않는 데이터는 다시 복구할 수 없었음)

- GA4는 필터로 인해 보고서에 보이지 않았던 데이터가, 언제든지 원상 복구 가능

<br>

- 기여 모델 (Attribution Model) : 캠페인을 통해 발생한 각각의 터치 포인트가 전환에 얼마나 기여했는지를 평가하는 것
  - GA4에서는 기본적으로 '데이터 기반 기여 모델'을 차용

<br>

### 유연해진 보고서
- GA4에는 맞춤 보고서 기능을 포함하고 있는 '탐색' 기능이 존재
  - 일반 적인 table 형태 뿐만 아닌, 자유 형식, 유입경로, 탐색 분석 등 다양한 형태의 보고서 제공

<br>

## 04. 구글 애널리틱스4 필수 용어
- **'보고서의 형태를 볼 때 알아야 하는 용어'**와, **'보고서의 내용을 볼 때 알아야 하는 용어'**
  - 전자에 해당하는 용어들은 '측정기준', '측정항목', '세그먼트', '필터' 등
  - 후자에 해당하는 용어들은 '세션', '참여', '평균 참여 시간', '이벤트', '주요 이벤트 (전환 등)'

<br>

### 04-01. 보고서의 형태를 볼 때 알아야 하는 용어

<br>

#### 1. 측정기준 (Dimension)
- **측정기준은 데이터를 어떤 기준으로 나누어 볼 것인지에 대한 기준**
- 데이터는 본래 하나의 뭉텅이이며, 이 뭉텅이를 필요에 따라 국가, 연령 단위로 나누어 볼 수 있음
  
  이 때, **'국가', '연령'과 같이 데이터 뭉텅이를 나누어 보는 기준이 측정기준**

  - 국가라는 측정기준으로 나누게 되면, **'한국', '미국' 등이 측정기준 값**

<br>

#### 2. 측정항목 (Metric)
- **측정항목은 뭉텅이 데이터에서 '내가 보고자 하는 부분'**
  - 예) 국가별로 데이터를 나누었는데, 나누어진 상태에서 '어떤 것을 볼 것인가'라는 것이 측정 항목
  - 국가별 '사용자 수', '전환 수' $\rightarrow$ 측정항목
  - GA4에서 대표적인 측정항목은 사용자 수, 세션 수, 페이지 뷰 수, 이벤트 수 등

<br>

<br>

#### 3. 세그먼트
- 사용자 기반 데이터를 특정 기준에 따라 분류하는 기능
- GA4는 유연한 사용자 분석을 위해, 세그먼트 기능을 활용하여 사용자의 행동, 특성, 구매 패턴 등 다양한 측면에서 데이터를 분석 가능
  - 세그먼트 기능을 활용하면, 남성 사용자와 여성 사용자로 분류하여, 그룹별로 어떤 매체를 통해 들어온 사용자들이 많았는지 등을 확인 가능
- GA4에서 세그먼트는 크게 3가지로 구분
1. 사용자 세그먼트 (User Segment)
2. 세션 세그먼트 (Session Segment)
3. 이벤트 세그먼트 (Event Segment)

<br>

#### 4. 필터
- 내가 보고자 하는 조건의 데이터만 보는 기능
  - 예) 사이트에 유입되는 사용자들 중 한국 사람들에 대한 데이터만 보겠다고 하면, 한국 사람만 포함하는 필터를 보고서에 적용
  
<br>

### 04-02. 보고서의 내용을 볼 때 알아야 하는 용어

<br>

#### 1. 세션
- 구글이 생각하는 유의미한 방문
- GA4에서 세션을 초기화하는 데 몇 가지 기준이 존재하며, 
  
  이 중 가장 중요한 기준은 사용자가 우리 사이트에서 더 이상 상호작용을 일으키지 않는 시간
- 구글 애널리틱스는 어떤 사람이 홈페이지나 앱에 접속하고 사이트 내에서 활동하다가 30분 동안 아무런 상호작용이 없으면 세션을 종료
  
  $\rightarrow$ 새롭게 상호작용이 발생하면 새로운 세션으로 다시 집계

  - 예) A라는 사람이 홈페이지에 들어와 쇼핑 중
    
    $\rightarrow$  이 때 홈페이지는 1명이 접속했지만, 밥을 먹는 1시간 동안 홈페이지에서 아무런 상호작용이 없음

    $\rightarrow$ GA4는 '밥 먹기 전 활동' 세션 1, '밥 먹은 후 활동' 세션 2로 나누어 세션을 집계

    $\rightarrow$ A의 활동에 대해 GA4는 '사용자' - 세션1:세션2 로 기록


<img src='image/GTM/0101.jpg' width='600'>

- **세션 초기화 조건 시간은 30분을 기본값으로 함**
  - 비즈니스 특성에 따라 (OTT 등) 상호작용이 없는 것이 당연한 경우도 존재 $\rightarrow$ **세션 초기화 조건은 최소 5분부터, 최대 7시간 55분까지 5분 단위로 설정 가능**

<br>

#### 2. 참여
- 사이트에서 적극적으로 활동한 경우
  - **'적극적' : 활성 페이지 체류 시간이 10초 이상 지속, 전환 이벤트가 발생, 페이지 조회 또는 화면 조회가 2회 이상 발생한 경우에 참여로 집계**
  - 예) 사이트에 접속하는 사람들의 참여율이 10%라고 하면, 10명 중 1명만 사이트에서 적극적으로 활동하고, 나머지 9명은 거의 바로 이탈
- **참여로 인정하는 체류 시간의 기본 값은 10초**
  - 필요에 따라 참여로 인정하는 최소 체류 시간은 최대 60초 까지 10초 단위로 설정 가능

<br>

- 참여율은 전환율과 대게 유의미한 상관관계를 가지며, 참영율이 높아지면 사용자들의 전환 (구매, 회원가입 등)이 발생할 확률이 동시에 증가
  - **사이트에서 전환 수치가 환벽하게 집계 되고 있지 않다면, '참여'를 캠페인 성과 개선의 보조 기준으로 삼는 것이 하나의 전략**

<br>

#### 3. 참여 시간
- 참여율과 함께 전환과 높은 상관관계가 있는 중요한 보조 지표
- 참여 시간은 포그라운드(foreground) 상태의 웹페이지나 앱 화면에서 보내는 시간
  - 포그라운드는 실제 내가 보고 있는 화면에서 출력이 되는 상태 $\leftrightarrow$ 백그라운드 (background)
  - 앱 환경에서는 A 앱을 보다가 홈 버튼을 눌러, 내 화면에서 안 보이게 하고, 다른 앱을 활용하는 경우
  
  **내 화면에 보이는 앱은 포그라운드(foreground) 상태, 화면에 보이지는 않지만 앱이 작동하고 있다면 백그라운드(background) 상태**

<br>

- A라는 사이트에 접속해서 콘텐츠를 5분을 보고 A 페이지를 그대로 둔 상태에서
  
  $\rightarrow$ 새로운 탭을 열어 B 사이트를 접속해 10분간 구경

  $\rightarrow$ 그리고 전체 웹 브라우저를 OFF

  $\rightarrow$ **A 사이트에 대한 총 체류 시간은 15분, 참여 시간은 포그라운드 상태인 5분**

<br>

#### 4. 이벤트
- 이벤트는 GA4에서 매우 중요한 개념
- **이벤트라는 용어는 웹이나 앱에서 발생한 상호작용을 의미**
  - 하지만 모든 행동이 이벤트로 자동 수집되지는 않으며, 보고자 하는 특정 행동이 있다면 별도로 이벤트를 세팅해야 함
  
<br>

- **GA4를 사용하는 마케터 입장에서 이벤트는 '내가 보고자 하는 사용자의 행동'**
  - 구매 완료, 상세 페이지 조회, 회원 가입 등
  
  $\rightarrow$ 특정한 행동이 일어났을 때의 데이터를 수집하려면, 이벤트가 필수적으로 설정이 되어 있어야 함

  $\rightarrow$ **이벤트를 설정하기 위해서는 내가 원하는 시점에 발동이 되는 '이벤트 태그 (Event Tag)'를 설치 해야 함**

  ```javascript
  gtag('event', 'screen_view', {
    'app_name':'myAppName',
    'screen_name':'Home'
  });
  ```

  - 이벤트의 이름은 `screen_view`
    - **`screen_view`라는 이벤트가 발생하였다고 하였을 때, 이 이벤트는 `myAppName`이라는 앱의 `Home`화면에서 발생한 이벤트 임을 의미**
    - **여기서 `app_name`과 `myAppName` 부분은 합쳐서 매개변수 (parameter)로 정의**
    - 이 매개변수를 한번 더 분류하면 `app_name`부분은 `key`로, `myAppName`은 `value`로 구분

<br>

- **GA4 이벤트는 구현 방법과 개발자가 설정하는 이벤트 명에 따라 총 4가지 종류로 분류**

| 구분 | 이벤트 명 | 설명  | 예 |
| - | - | - | - |
| 자동 | 자동 수집 이벤트 | 기본 데이터 수집을 통해 자동으로 수집되는 이벤트 | `page_view`, `session_start` 등 |
| 반자동 | 향상된 측정 이벤트 | 구글에서 제공하는 옵션을 통해 추가적으로 수집할 수 있는 자동 이벤트 | `scroll`, `view_search_result` 등 |
| 수동 | 추천 이벤트 | 구글에서 사전 정의된 이벤트 및 매개변수명을 활용하여 설치하는 이벤트 | `add_to_cart`, `purchase` 등 |
| 수동 | 맞춤 이벤트 | 이벤트와 매개변수명을 원하는 이름으로 커스텀하여 설치하는 이벤트 | `N_pay_click`, `purchase_complete` 등 |

<br>

**1) 자동 수집 이벤트**
- GA4를 설치하면 별도의 작업 없이 자동으로 수집되는 이벤트
  - 예) 웹 페이지 방문, 앱의 첫 실행, 세션 시작 등

<br>

**2) 향상된 측정 이벤트**
- 스크롤, 아웃바운드 클릭, 사이트 내 검색, 비디오 조회 등 추가적인 사용자 상호작용을 자동으로 추적

<img src='image/GTM/0102.png' width='500'>

- 향상된 측정 이벤트의 전제 조건:
  - **향상된 측정 이벤트 데이터가 GA4에 정상적으로 수집되려면, 기존 웹사이트나 앱에서 해당 이벤트가 GA4에서 인식할 수 있는 조건으로 설정되어 있어야 함**
  - 예) 사이트 내부 검색어 데이터를 얻고자 할 때, GA4가 기본적으로 수집하는 사이트 검색 매개변수는 `q`, `s`, `search`, `query`, `keyword` 5가지
    
    **만약 사이트가 기본값으로 수집하는 위 매개변수와 다른 이름으로 내부 검색어를 수집하고 있다면, GA4에서 내 매개변수를 추가하거나 GA4에서 인식할 수 있는 매개변수로 이를 변경해 주어야 함**

<br>

**3) 추천 이벤트**
- **구글은 특정 비즈니스 목적이나 앱 유형에 맞춰 추천하는 이벤트 목록을 제공 (예: '전자상거래')**
  - **`view_item`, `add_to_cart`, `begin_checkout`, `purchase`와 같은 이벤트만 인식되고, 제품에 대한 정보는 `item`이란 이름의 매개변숫값으로 지정해야 올바르게 데이터가 수집됨**
  - https://support.google.com/analytics/answer/12200568?hl=ko

<br>

**4) 맞춤 이벤트**
- 맞춤 이벤트는 구글에서 제안하는 추천 이벤트 외에 사용자가 직접 정의한 이벤트
- 원하는 타이밍에 원하는 이름으로 이벤트를 세팅하는 것이 맞춤 이벤트
- 추천 이벤트와 달리, 자동으로 연동되는 보고서가 없음
  - 만약 구매 완료 시점을 이벤트로 설정하고 싶은 경우, `purchase`로 명명하면 GA4에서 자동으로 이 이벤트 명을 인식하여 전자상거래 보고서 등에 연동이 되지만

    `purchase_complete`처럼 추천 목록에 없는 이름으로 명명하면, 같은 구매 완료를 뜻하는 이베트라고 하더라도 GA4가 구매 완료라는 의미를 바로 인식하지 못함

    $\rightarrow$ 이벤트를 기획할 때는 구글의 추천 이벤트 명과 매개변수 명을 우선적으로 사용하고, 그 이후에 필요에 따라 맞춤 이벤트 명을 설정하는 것이 중요

<br>

**5) 전환 (주요 이벤트)**
- 전환의 정식 명칭은 '주요 이벤트'
- GA4에서 설정한 일반 이벤트 중 비즈니스 목표에 적합하다고 판단되는 중요한 이벤트 데이터를 별도로 설정하는 것
- 어떤 이벤트를 주요 이벤트로 설정할지에 대해서는 GA4 사용자가 스스로 결정
- 하나의 속성당 최대 30개의 전환 이벤트를 설정할 수 있음
