#### 참여도 보고서에는 4가지 유형의 보고서가 존재
1. 개요

3. 전환수
4. 페이지 및 화면

### GA4 참여 - 이벤트 리포트

- GA4에서 이벤트란, 웹사이트나 애플리케이션에서 발생하는 특별한 사건
- 데이터 리포트의 측정 기준 이름은 이벤트 이름
  - `view_promotion`, `view_item_list`, `page_view`, `user_engagement`, `scroll`, `session_start` 등

### GA4 참여 - 전환
- **이벤트** : 우리 웹사이트나 어플리케이션에서 발생한 특별한 사건(ex. 장바구니 담기, 결제 완료, 스크롤, 영상 조회 등)
- **전환** : 우리가 추적하고 있는 이벤트들 중 특별히 중요한 사건(ex. 결제 완료)

### 페이지 및 화면 리포트
- 페이지 및 화면 리포트 내에 있는 측정 기준의 이름은 `[페이지 경로 및 화면 클래스]`
  
  여기서 페이지 경로란, 메인 도메인을 제외하고 웹사이트의 서브 도메인에 해당되는 영역을 의미

<img src='https://lh7-us.googleusercontent.com/F38R1WnSVJZw_24TvunTYyc54YaiI8SlZ_c3ib3K2ieMoAwgxk94jQ70g-_87QGaT0MT7GaCkTCJlvGvGWlTCk5bsNFLtpTE3hqDMOvauBeegnV3V3EmvT21inpDFC3kuMTht-hLtUW00hs8wgSCuZ4' width='500'>

<br>

#### 예: `www.example.com/event` 라는 페이지 URL이 있을 때,
- 여기서 `www.example.com`은 우리 사이트의 메인 도메인이니 단순히 구글애널리틱스에서는 슬래시[`/`] 하나로 표현 $\rightarrow$  **메인도메인은 생략**
- 뒤쪽에 오는 `/event`는 서브 도메인 $\rightarrow$ [페이지 경로] 측정 기준은 이러한 서브 도메인을 제공하는 측정 기준
- 웹사이트에는 각각의 페이지가 URL이 있는것처럼, [화면 클래스]는 어플리케이션의 화면 하나하나를 일컫는말이라고 이해 가능

### 페이지 및 화면 리포트의 다양한 측정 기준

1. **페이지 제목 및 화면 클래스**
-  **`[페이지 제목]` 측정 기준은 우리 웹사이트 내 개별 웹 페이지에 있는 제목값을 의미**
  - 구글 애널리틱스는 `<title>` 태그 안에 들어가는 텍스트 값을 그대로 가져오도록 설정
  - 만약에 타이틀 값이 모두 다 똑같은 웹사이트라면 한개의 행만 나타날 수 밖에 없음 $\rightarrow$ `[페이지 경로]`로 URL별로 데이터를 봐야하는 불편함 발생
  - 웹사이트를 만들때 기본적으로 `<title>` 태그 안에 들어가는 텍스트를 페이지별로 다르게 식별을 해주면 구글 애널리틱스에서도 데이터를 한글값으로 볼 수 있으며, 검색엔진에도 긍정적인 영향
   
  <img src='https://lh7-us.googleusercontent.com/nL5RRnBbqjy6VrOutnOlC_5XIYqhTW-fiks4oqXe9rHtsFJ3Y4esDlGEdRgxOIO0L3QQVKTnBxoCmpVPc1yLeQGcKKW1Cy0wL1WIoqHJxd753jVVmmgrSaAZca0LdlIlHCW8ttra7wWWG9bDsqZ7yWk' width='500'>

#### 페이지 리포트에서 주의해야할 점

1) 측정 항목이 `[사용자 수]` 또는 `[세션]`이 아니라 `[조회수]`
- 페이지 제목이든, 페이지 경로든 페이지 단위로 데이터를 볼 때에는 한명의 유저가 여러 페이지를 조회할 수 있기 때문에, 페이지별로 사용자수를 보고할 수가 없음

<img src='https://lh7-us.googleusercontent.com/QjboW8c22QopaUgONaDZ5IlHK2m9zG57leLXZ-sz36I3Cg6NKOD7l1Ffl5vkULLEvPI92eBl6h2aT32gSCeJzkfSmlM-MXLBE4vBGIcevQNII8Gv6OZnBkAByNxo96DqPHbqOFVorUf2sBbqNM7Zw68' width='500'>

<br>

2) 페이지별로 이벤트나 전환이 얼마나 발생했는가를 보는것은 많은 인사이트를 주지 않음'
- 만약에 우리가 쇼핑몰을 운영하고 있고, 결제 완료 페이지가 있다면 당연히 [Purchase] 이벤트는 결제 완료 페이지에만 발생하기 때문에, 이벤트가 해당 페이지에서만 발생
  
  $\rightarrow$ 그렇기 때문에 특정 페이지별로 전환율을 보는것은 무의미

  $\rightarrow$ `[비교 기능]`을 사용하여 페이지 조회수의 변화 추이를 관측

### GA4 방문 페이지 리포트
- **방문 페이지는 말 그대로, 웹사이트에 고객이 들어오는 순간에, 방문했던 페이지의 URL을 의미**
  
  $\rightarrow$ 방문 페이지는 1개의 세션당 1개밖에 없을 것

  $\rightarrow$ 방문 페이지 리포트는 페이지 리포트랑은 달리 `[세션 수]` 부터 시작

#### GA4 방문 페이지의 인사이트
- 결국엔 “고객을 어디로 랜딩시켜야, 최종 구매로 이어지는 전환율이 높을까?
  - 예) 제품이 여러개 진열되어 있는 [카테고리 페이지]와 제품 1개의 대한 상세 정보만 기록되어있는 [상세페이지]가 있다고 했을 때, 
    
  방문 페이지별 거래 전환이 얼마나 발생했는지 관측하면서, 우리 사이트에는 고객이 어디로 방문하는것이 구매에 좋을까에 대한 인사이트를 얻을 수 있음

### 전환율의 함정

<img src='https://lh7-us.googleusercontent.com/SiJpy7Kyj8JNoMyN04S4_qqfcJTXgp8IPjR3_LhM1kNHkExiDrapj_mKIneGS4IXkfCuymFpK_r-To3pHnAiWXtyDvlugJE1BhTvdZ3ZMon47YU9WBhgNRENrt1Sk0FzeC__74_BCNtBMkP4MWPB6yE' width='500'>

> ”100명이 웹사이트에 들어와서 10명이 결제 완료까지 이어졌을 때 전환율이 10%”

- 분모에는 전체 방문자수가 들어가고, 분자에는 결제 완료수가 들어간 것
- 하지만 상품별로 볼 때는 분모 분자를 나누면 안 됨
  - A상품의 전환율을 관측하기 위해서는 분모에 A 상품을 조회한 횟수, 분자에는 A 상품을 거래한 횟수가 나와야하는게 올바름
  - 예) 제품의 상세페이지 조회수는 ‘4,778’건이고, 구매 수는 ‘403’ $\rightarrow$ 해당 제품의 전환율은 403/4778 $\rightarrow$ 전환율이 무려 8.4% $\rightarrow$$ 상품별로 전환율을 추출하여 해당 제품의 매력도를 파악하고 유추
  
  $\rightarrow$ **특정 제품을 단 한 명이 조회했는데, 그 한 명이 구매? $\rightarrow$ 전환율이 100%**

  $\rightarrow$ **전환율과 같은 비율 데이터에는 [분모]에 가중치를 둬서 데이터를 관측해야 함**
