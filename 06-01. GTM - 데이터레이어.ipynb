# 06. 데이터레이어 (Data Layer)
- **데이터레이어는 웹사이트에서 태그 매니저 컨테이너로 정보를 전달하는 매개체 역할**
    
    웹사이트에 GTM이 설치되어 있다면, 바로 데이터레이어를 활용할 수 있음

- 구글 공식 문서에 따르면, 데이터 레이어란 웹사이트에서 태그 매니저 컨테이너로 정보를 전달할 때 사용되는 자바스크립트 배열
    
    $\rightarrow$ 전달된 정보는 태그 구성 시 변수를 게재하거나 트리거를 활성화하는 데 사용

- 태그 관리자는 변수, 거래 정보, 페이지 카테고리 및 페이지 곳곳에 존재하는 기타 중요 신호를 참조하는 대신 데이터 영역 소스 코드에 있는 정보를 쉽게 참조하도록 설계
- 데이터 영역은 변수를 비롯한 연결된 값을 사용해 구현되기 때문에 태그를 실행해야 할 때 변수와 값을 바로 사용할 수 있음

    <img src="https://lh7-rt.googleusercontent.com/docsz/AD_4nXchwh7TZz1K-Asn1YIPgJoKCgyiucfLWCTGy3pZjh8P7gE9_JUGGFaBWr-s3DU-CwLpqVm9HfNtsyC1ugJ0Bi9wUR6solfB4tSzYTTMeL_L0wjx3UOV0QHCFUQBgMR607db9rWhAnmDebnHNuXLZR-Rd5D1?key=txnZy4deI_ikSgI5kjB2pg" width="400">

- 웹사이트에서 사용자가 버튼 클릭 또는 페이지 조회와 같은 활동을 할 때마다 데이터레이어로 활동에 대한 데이터들이 전달되고 
  
  전달된 데이터를 활용하여 GTM에서 트리거 또는 변수로 활용하여 이벤트 데이터를 GA4로 전송

- 웹사이트에서 사용자가 활동을 할 때, 자동으로 데이터레이어로 전달되는 데이터도 있고, 특정 시점에 동적 변수를 넣고 싶을 때에는 간단한 데이터레이어 코드를 삽입해주면 GTM 에서 이를 감지하고 변수를 활용
  
<br>

#### 데이터레이어의 활용
1. GTM 트리거로 잡을 수 없는 시점의 이벤트를 GA4로 전송해야 할 때
2. GTM에서 GA4로 이벤트를 전송할 때 동적인 이벤트 매개변수를 넣기 위해서
    - 예) 상품 상세페이지 조회 시점에 이벤트만 보내는 것 보다 상품에 대한 추가적인 정보들을 같이 보낸다면 GA4에서 더욱 데이터를 상세하게 분석하는 것이 가능
    - 정적인 이벤트 매개변수는 GTM에서 상수 값으로 보낼 수 있지만, 동적인 이벤트 매개변수는 데이터레이어를 통해서 매개변수에 대한 데이터가 전달되어야만 해당 매개변수를 GTM에서 GA4로 보낼 수 있기 때문

### 모든 데이터레이어 작업을 웹사이트 소스 코드 내부에 삽입하지 않고 GTM에서 할 수 있을까?
- **모든 데이터레이어 작업을 웹사이트 소스 코드를 내부에 삽입하지 않고 GTM에서 작업하는 것은 불가능**

<br>

- 예시 1 : 구매완료 페이지에서 상품 카테고리 값을 매개변수로 가져와야 하는데, 값을 서버에서 관리하고 있고 화면에서는 보이지 않는 경우:
    
    $\rightarrow$ GTM에서 매개변수를 가지고 올 때는, 화면에 보이는 값만 가지고 올 수 있음

<br>

- 예시 2 : alert창 조회 시점으로 이벤트를 잡는 경우:
  
    $\rightarrow$ alert창 조회 또는 클릭은 GTM 트리거로 잡는 것이 불가능

<br>

- 위 2가지 예시처럼 GTM 트리거로 잡기 어려운 이벤트이거나 화면에서 보이지 않는 매개변수를 가지고 와야 하는 경우는, 
  
  GTM에서 데이터레이어 작업이 불가하므로 웹사이트 내부 소스 코드 작업이 필요
